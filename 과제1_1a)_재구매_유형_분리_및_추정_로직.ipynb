# 03) 고객 관련 정보 추가

### a-1) TYMD(생년월일) 컬럼 이용해서 AGE(연령) 컬럼을 추가하고, AGE를 범주화한 AGE_R 컬럼 추가
### 영유아, 아동, 청소년은 면허 취득 가능 나이 제한 및 구입 능력이 없다고 판단하여 제외할 예정 
> 청년 (19-29) / 중년 (30-49)/ 장년 (50-64) / 노년 (65세 이상)
### a-2) TYMD(생년월일) 컬럼 이용해서 SEASON(계절) 컬럼 추가
> 3-5월: 봄 / 6-8월: 여름 / 9-11월: 가을 / 12-2월: 겨울

```python
AGE = dict()
AGE_R = dict()
SEASON = dict()

for i in range(len(new_df1)):
    TYMD = new_df1['TYMD'][i]
    TYMD = str(TYMD)
    TYMD = TYMD[0:4]
    TYMD = int(TYMD)
    
    date = new_df1['TYMD'][i]
    date = str(date)
    date = date[4:6]
    date = int(date)
    
    age = 2021 - TYMD + 1
    
    AGE[new_df1['CUS_ID'][i]] = age

    if 19 <= age <=29:
        AGE_R[new_df1['CUS_ID'][i]] = '청년'
    elif 30 <= age <=49:
        AGE_R[new_df1['CUS_ID'][i]] = '중년'
    elif 50 <= age <=64:
        AGE_R[new_df1['CUS_ID'][i]] = '장년'
    elif age >= 65:
        AGE_R[new_df1['CUS_ID'][i]] = '노년'
    else:
        AGE_R[new_df1['CUS_ID'][i]] = '해당없음'
    
    
    if 3 <= date <=5:
        SEASON[new_df1['CUS_ID'][i]] = "봄"
    elif 6 <= date <=8:
        SEASON[new_df1['CUS_ID'][i]] = "여름"
    elif 9 <= date <=11:
        SEASON[new_df1['CUS_ID'][i]] = "가을"
    else:
        SEASON[new_df1['CUS_ID'][i]] = "겨울"
```

```python
AGE = pd.DataFrame(list(AGE.items()),
                   columns=['CUS_ID', 'AGE'])

AGE_R = pd.DataFrame(list(AGE_R.items()),
                   columns=['CUS_ID', 'AGE_R'])

SEASON = pd.DataFrame(list(SEASON.items()),
                   columns=['CUS_ID', 'SEASON'])
```

### AGE 컬럼에서 연령이 19세 미만인 행 제외

```python
del_id = []
for i in range(len(AGE['AGE'])):
    if AGE['AGE'][i] < 19:
        del_id.append(AGE['CUS_ID'][i])
        print(AGE['CUS_ID'][i],":",AGE['AGE'][i])
print(len(del_id))       
```

### 도출된 306개의 데이터는 삭제하고 진행

```python
mask = AGE['CUS_ID'].isin(del_id) 
AGE = AGE[~mask]
```

```python
mask = AGE_R['CUS_ID'].isin(del_id) 
AGE_R = AGE_R[~mask]
```

```python
mask = SEASON['CUS_ID'].isin(del_id) 
SEASON = SEASON[~mask]
```

```python
mask = new_df1['CUS_ID'].isin(del_id) 
new_df1 = new_df1[~mask]
```

```python
mask = new_df2['CUS_ID'].isin(del_id) 
new_df2 = new_df2[~mask]
```

```python
new_df1 = pd.merge(new_df1, AGE, how='outer', on='CUS_ID')
```

```python
new_df1 = pd.merge(new_df1, AGE_R, how='outer', on='CUS_ID')
new_df1 = pd.merge(new_df1, SEASON, how='outer', on='CUS_ID')
```

## a-1) 차량 보유 기간과 관련된 컬럼 기준 확립
###  - 기본적으로 [ 차량보유기간 = 보유종료일자 - 보유시작일자(출고일자) ] 로 계산
###  - '출고일자'와 '보유시작일자' 컬럼 중 어떤 것을 사용할 것인지 결정하기 위해 출고일자!=보유시작일자 인 값들에 대해 
###  날짜 차이를 구해보고 특별한 문제가 없다면 '출고일자'를 사용하겠다.
#### (보유시작일자보다 출고일자가 차량이 필요해서 구입한 날짜를 가장 잘 반영한다고 판단됨)

```python
check = sample_car[sample_car['WHOT_DT']!=sample_car['CAR_HLDG_STRT_DT']] # 출고일자 != 보유시작일자 인 데이터
check= check.reset_index(drop=True)
check.head()
```

```python
len(check)
```

```python
from dateutil.parser import parse 

result = dict();

for i in range(len(check)):
        diff = ''
        diff = parse(str(check['CAR_HLDG_STRT_DT'][i])) - parse(str(check['WHOT_DT'][i]))
        result[check['CUS_ID'][i]] = int(diff.days)
```

### 차량 출고 이전에 미리 등록하는 경우 '보유시작일자' - '출고일자' 가 음수인 상황은 표본에는 없는 것으로 보임

```python
for i in result.values():
    if i<0:
        print(i)
```

### 그러나 출고 후 임시 번호판 주행 가능한 날짜가 10일이내인 것을 고려할 때, 아래의 9건의 데이터에는 문제가 있는 것으로 보여진다.

```python
cnt=0
id = list(result.keys())
for i in id:
    if result[i] > 10:
        cnt+=1
        print(i,": ",result[i])
print("count =",cnt)
```

### 예를들어 'CUS_ID'가ATZOHZV310000245인 사람의 경우 
### 출고일자: 2013년 12월 27일 / 보유시작일자: 2015년 10월 20일 / 보유종료일자: 현재 보유중 으로
### '출고일자'와 보유시작일자' 데이터에 엄청난 차이가 존재한다. 

```python
check[check['CUS_ID']=='ATZOHZV310000245']
```

### 보유종료일자 - 출고일자 < 0 인 오류가 없다면 '보유시작일자'대신 '출고일자' 데이터를 사용하겠다.

### 재구매 유형 분리 기준을 세우기 위해 가장 많은 값의 분포를 지닌 'CCAR_HLDG2_1' 컬럼을 분석
### 모든 값들을 절댓값 처리하여 양수로 만들고 평균, 백분위수, 최댓값, 최솟값 등 확인
### 적정한 대푯값을 선정하여 그 값을 기준으로 '대차'와 '신차구매' 를 구분

```python
Abs = abs(sample[sample['CAR_HLDG2_1'].isna()==False]['CAR_HLDG2_1'])
Abs
```

```python
print(max(Abs)) # 최대값
print(min(Abs)) # 최소값
print(np.mean(Abs)) # 평균값
print(Abs.quantile(.25)) # 1사분위 수
print(Abs.quantile(.5)) # 2사분위 수
print(Abs.quantile(.75)) # 3사분위 수
```

```python
plt.boxplot(Abs)
```

### - 최댓값과 최솟값의 차이가 큰 것으로 보아 자료의 분포가 넓게 퍼져있다는 것을 유추할 수 있다.
### - 그렇기 때문에 일반적인 산술 평균값을 대표값으로 선정하기에는 무리가 있다.
### - 사분위 수 중  Q2(2사분위 수; 중앙값) 또는 가중평균 중 더 적당한 것을 대푯값으로 선정하겠다.
### - ▲ 가중평균은 대차가 보통 빠른 시일내에 이루어진다는 점에서 다음과 같이 계산하겠다. 
#### 가중 평균 = { (가중치*0-90) + (가중치*91-180) + (가중치*181-365) + (가중치*367-730) + (그 외) } / N (N: 자료의 총 개수)
>각각의 가중치는 0-90일: 1 / 91-180: 0.5 / 181-365: 0.3333 / 367-730: 0.25 / 그 외: 0.2 이다. 
>> 5점 척도에서 0-90을 5점으로 즉, 가중치를 1로 두고 나머지는 4점(1/2), 3점(1/3), 2점(1/4), 1점(1/5)로 두었다.

```python
Sum = 0;
N = len(Abs);

for i in Abs:
    if 0 <= i <= 90:
        Sum += i
    elif 91 <= i <= 180:
        Sum += i*0.5
    elif 181 <= i <=365:
        Sum += i*0.3333
    elif 367 <= i <= 730:
        Sum += i*0.25
    else:
        Sum += i*0.2
print("가중평균 =",Sum/N)
```
